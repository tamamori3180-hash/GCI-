## 目標
　Titanic号の乗客に関する様々な情報から、**その乗客が最終的に亡くなったかどうか**をできるだけ正しく予測してください。乗客の死亡/生存情報は、Perishedカラムの値に対応付けられています。Perishedカラムの値が1であれば死亡した、0であれば生存したということに対応しています。本コンペはオリジナルの"Titanic: Machine Learning from Disaster"コンペを模してはいるものの同一ではないため、一部カラムの命名や意味が異なります。下記のデータの項をよく読んで取り組むようにしてください。

## データ
　データ分析コンペティションでは、一般的に**訓練データ**と**テストデータ**が与えられます。今回の場合、訓練データは**data/train.csv**、テストデータは**data/test.csv**となっています。

訓練データとテストデータは以下のような点で異なります。  
- 訓練データ：このデータによって機械学習モデルを学習させます。
- テストデータ：このデータに対する予測を提出して、その精度を競います。

Nameに含まれる'Miss'と'Mrs'という敬称で平均年齢にズレがあるかも

```python
def extract_title(name):
    """名前から敬称（Mr, Miss, Mrsなど）を抽出する"""
    title_search = name.split(', ')[1].split('.')[0].strip()
    return title_search

# 1. Nameカラムから敬称（タイトル）を抽出し、新しいカラム'Title'を作成
df['Title'] = df['Name'].apply(extract_title)

# 2. 'Miss'と'Mrs'のみをフィルタリングし、平均年齢を計算
average_ages = df[df['Title'].isin(['Miss', 'Mrs'])].groupby('Title')['Age'].mean()

# 3. 結果を表示
print(average_ages)
```

敬称としての「Master」
「Mr.」の由来: 過去には、社会的な一人前の男性に「Mr.」が使われる以前は、「master」が敬称として使われていました。

```python
# 3. 'Title' ごとに中央値を算出し、その値で Age の NaN を補完
# groupby().transform('median') が最も効率的で推奨される方法です。

# グループごとの中央値を計算し、それを Age の NaN 部分に代入
dfn['Age'].fillna(dfn.groupby('Title')['Age'].transform('median'), inplace=True)

# 補完後の Age の欠損値の数を確認（0になっているはずです）
print("Ageカラムの欠損値補完後の残りの数:", dfn['Age'].isna().sum())
```

```python
import numpy as np                                   # 数値計算や配列操作
import pandas as pd                                  # 表形式のデータを扱う
import matplotlib.pyplot as plt                      # 基本的なグラフ描画
import seaborn as sns                                # きれいで便利な統計グラフ
from sklearn.preprocessing import LabelEncoder       # カテゴリ変数を数値に変換
from sklearn.ensemble import RandomForestClassifier  # ランダムフォレスト分類器
from sklearn.model_selection import train_test_split # データの分割
from sklearn.metrics import accuracy_score
```

```python
# 特徴量と目的変数に分ける
X = dfn3.drop(columns=["Perished"])  # 予測に使う説明変数
y = dfn3["Perished"]                 # 予測したい目的変数

# データを7:3に分割（訓練データ70%、検証データ30%）
X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.3, random_state=46, stratify=y
)

# モデルの設定（ランダムフォレスト）
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    random_state=2025
)

# モデルを訓練データで学習
model.fit(X_train, y_train)

# 検証データで予測（クラスラベルを出力）
y_valid_pred = model.predict(X_valid)

# Accuracyスコアで性能を評価
accuracy = accuracy_score(y_valid, y_valid_pred)
print("検証データの正解率:", round(accuracy, 4))
```
